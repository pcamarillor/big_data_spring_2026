```python
import findspark #it's going to look for an existing installation that we've already have

findspark.init()

from pyspark.sql import SparkSession


spark = SparkSession.builder \
        .appName("1st SparkSession") \
        .master("spark://spark-master:7077") \
        .config("spark.ui.port","4040") \
        .getOrCreate()

#create SparkContext
sc = spark.sparkContext
```

```python

data = [
    ('apple', 1),
    ('banana', 2),
    ('apple', 3),
    ('orange', 4),
    ('banana', 1),
    ('orange',2)
]

rdd_data = sc.parallelize(data)
result = rdd_data.reduceByKey(lambda a, b: a + b)
r = result.collect()
r


```

```python
groupped_rdd = rdd_data.groupByKey()
group = groupped_rdd.collect()
for k, v in group:
    print(f"k:{k}: {list(v)}")
```

```python
words = ["apple", "banana", "apple", "orange",
"banana", "apple", "orange", "orange"]
words_rdd = sc.parallelize(words)

counter = words_rdd.countByValue()
counter
```
