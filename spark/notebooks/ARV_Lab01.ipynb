```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

```python
print(collaborations)
```

```python
unique_students_id = set()
for i in collaborations:
    for j in i:
        unique_students_id.add(j)

print(unique_students_id)
```

```python
pair_count = {}
for record in collaborations:
    for i in range(len(record)):
        for j in range(i + 1, len(record)):
            temp_list = [record[i], record[j]]
            temp_list.sort()
            pair = tuple(temp_list)
    
            if pair in pair_count:
                pair_count[pair] += 1
            else:
                pair_count[pair] = 1

print(pair_count)
```
