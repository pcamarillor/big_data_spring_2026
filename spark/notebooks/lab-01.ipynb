# Lab 01: Student Collaboration Analyzer

```python
from itertools import combinations

collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

### Task 2: Unique Students (Sets)

```python
students_set = set()
```

```python
for collaboration in collaborations:
    for student in collaboration: students_set.add(student)
    # end for-in
# end for-in
```

```python
students_set
```

### Task 3: Collaboration Frequency (Dictionaries)

```python
frequencies = {}
```

```python
for collaboration in collaborations:
    for pair in combinations(collaboration, 2):
        pair = tuple(sorted(pair))
        
        if pair in frequencies: frequencies[pair] += 1
        else: frequencies[pair] = 1
        # end if-else
    # end for-in
# end for-in
```
