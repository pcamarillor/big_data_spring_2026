Task 3

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

collaboration_count = {}

for group in collaborations:
    for i in range(len(group)):
        for j in range(i + 1, len(group)):
            student1 = group[i]
            student2 = group[j]

            pair = tuple(sorted((student1, student2)))

            if pair in collaboration_count:
                collaboration_count[pair] += 1
            else:
                collaboration_count[pair] = 1

print(collaboration_count)
```
