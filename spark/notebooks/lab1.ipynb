```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

uniqueNumber=set()

for limpia in collaborations:
    for inside in limpia:
        uniqueNumber.add(inside)

print(uniqueNumber)
```

```python
diccionario={}
for sort in collaborations:
    for i in range(len(sort)):
        for j in range(i+1,len(sort)):
            a=sort[i]
            b=sort[j]
            pair=tuple(sorted([a,b]))
            if pair in diccionario:
                diccionario[pair]+=1
            else:
                diccionario[pair]=1
print(diccionario)
```
