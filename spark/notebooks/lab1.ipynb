Lab 01: Student Collaboration Analyzer

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

```python
studentsSet = set()
for studentsGroup in collaborations:
    studentsSet.update(studentsGroup)
print(studentsSet)
```

```python
def pair(a,b):
    return tuple(sorted((a, b)))

def createPairs(group,frequency,functionPair):
    l = len(group)
    for i in range(l):
        for j in range(i+1,l):
            key = functionPair(group[i],group[j])
            frequency[key] = frequency.get(key,0) + 1

collaborationFrequency = dict()

for studentsGroup in collaborations:
    createPairs(studentsGroup,collaborationFrequency,pair)

print(collaborationFrequency)
        
        
```
