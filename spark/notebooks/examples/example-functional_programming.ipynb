### Example of a pure function in python

```python
def add(x, y):
  return x + y

print(add(18, 193.4))
```

### Example of Pure Functions: Calculate Factorial

```python
def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)

print(factorial(5))
```

### Functions as first-class citizens

```python
def greet(name):
  return f"Hello, {name}!"
  
# Assigning the function to a variable
say_hello = greet
  
# Using the new variable to call the function
print(say_hello("Alice")) 
```

### Functios as Arguments

```python
def apply_function(func, value):
    return func(value)
  
def square(x):
    return x * x
  
# Passing the function `square` as an argument to 
# `apply_function`
result = apply_function(square, 5)
print(result)  # Output: 25
```

### Functions in Data Structures

```python
def add(x, y):
  return x + y
  
def subtract(x, y):
  return x - y
  
# Store functions in a dictionary
operations = {
  'add': add,
  'subtract': subtract
}
  
# Use the dictionary to call functions
print(operations['add'](10, 5))      # Output: 15
print(operations['subtract'](10, 5)) # Output: 5
```

### Variable Arguments Using *args

```python
def sum_numbers(*args):
  total = 0
  for num in args:
      total += num
  return total

print(sum_numbers(1, 2, 3))        # Output: 6
print(sum_numbers(10, 20, 30, 40))
```

```python
def interate_args(*args):
    for a in args:
        print(f"Type of {a} is {a.__class__}")

interate_args("hi", 77, 82.5)
```

### Lambda expressions

```python
# Regular function
def square(x):
  return x ** 2

# Equivalent lambda function
square_lambda = lambda x: x ** 2

print(square(5))

print(square_lambda(6))
```

### Map function

```python
nums = [1, 2, 3, 4, 5]

# Example 1: square each number
squares = list(map(lambda x: x*x, nums))
print(squares)
# Example 2: normalize strings (pure transformation)
names = ["  rObert", "ALICE  ", "bob"]
clean = list(map(lambda s: s.strip().title(), names))
clean
```

### Filter

```python
# Example 1: keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)

# Example 2: keep "valid" events (predicate)
events = [{"id": 1, "ok": True}, {"id": 2, "ok": False}, {"id": 3, "ok": True}]
valid = list(filter(lambda e: e["ok"], events))
print(valid)
```

### Sorted

```python
# Example 1: sort strings by length (key function)
words = ["spark", "kafka", "aws", "emr"]
by_len = sorted(words, key=lambda w: len(w))
print(by_len)

# Example 2: sort dicts by a field (key extraction)
movies = [
  {"title": "A", "rating": 7.2},
  {"title": "B", "rating": 8.1},
  {"title": "C", "rating": 6.9},
]
top_first = sorted(movies, key=lambda m: m["rating"], reverse=True)
print(top_first)
```
