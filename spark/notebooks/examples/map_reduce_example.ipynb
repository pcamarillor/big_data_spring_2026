```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
            .appName("Map-Reduce-IMG") \
            .master("local[*]") \
            .config("spark.ui.port", "4040") \
            .getOrCreate()

sc = spark.sparkContext

filenames = ["/opt/spark/work-dir/notebooks/img/ITESOLogo.png", "/opt/spark/work-dir/notebooks/img/text1.png", "/opt/spark/work-dir/notebooks/img/text2.jpeg"]

rdd = sc.parallelize(filenames)
```

```python
from session7.image_utils import extract_text
text_rdd = rdd.map(extract_text)
```

```python
text_lst = text_rdd.collect()
text_lst
```
