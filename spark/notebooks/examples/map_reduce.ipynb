```python
import findspark
findspark.init()


from pyspark.sql import SparkSession


spark = SparkSession\
    .builder\
    .appName("Hello, Spark")\
    .master("spark://spark-master:7077")\
    .config("spark.ui.port", "4040")\
    .getOrCreate()

sc = spark.sparkContext
```

```python
numbers  = [1, 43, 5, 24]
rdd = sc.parallelize(numbers)

def squares(n):
    return n ** 2

squares_rdd = rdd.map(squares)
r = squares_rdd.collect()
r
```

```python
data = [
('apple', 1),
('banana', 2),
('apple', 3),
('orange', 4),
('banana', 1),
('orange', 2)
]
rdd = sc.parallelize(data)
reduced_rdd = rdd.reduceByKey(lambda a, b: a + b)

r = reduced_rdd.collect()
r
```

```python
grupped_rdd = rdd.groupByKey()
group = grupped_rdd.collect()
for k, v in group:
    print(f"k:{k}: {list(v)}")
```

```python
words = ["apple", "banana", "apple", "orange", "banana", "apple", "orange", "orange"]

words_rdd = sc.parallelize(words)

word_counter = words_rdd.countByValue()

word_counter
```
