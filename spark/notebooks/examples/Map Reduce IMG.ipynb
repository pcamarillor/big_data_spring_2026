```python
from PIL import Image
import pytesseract

def extract_text(filename):
    img = Image.open(filename)
    return pytesseract.image_to_string(img)
```

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Map-Reduce-IMG") \
    .master("local[*]") \
    .config("spark.ui.port", "4040") \
    .getOrCreate()

sc = spark.sparkContext

filenames = ("/opt/spark/work-dir/notebooks/img/Boom_comic.png",
            "/opt/spark/work-dir/notebooks/img/NoSmoking_signal.png",
            "/opt/spark/work-dir/notebooks/img/OK_comic.png",
            "/opt/spark/work-dir/notebooks/img/Stop_signal.jpg",
            "/opt/spark/work-dir/notebooks/img/Logo-ITESO-Principal.jpg")

rdd = sc.parallelize(filenames)
```

```python
img = Image.open("/opt/spark/work-dir/notebooks/img/Logo-ITESO-Principal.jpg")
text = pytesseract.image_to_string(img)
text
```

```python
text_rdd = rdd.map(extract_text)

text_lst = text_rdd.collect()

text_lst
```
