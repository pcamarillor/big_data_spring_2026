```python
from PIL import Image
import pytesseract

def extract_text(filename):
    img = Image.open(filename)
    return pytesseract.image_to_string(img)
```

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
            .appName("Map-Reduce-IMG") \
            .master("local[*]") \
            .config("spark.ui.port", "4040") \
            .getOrCreate()

sc = spark.sparkContext

filenames = ["/opt/spark/work-dir/notebooks/img/image1.png"
            ,"/opt/spark/work-dir/notebooks/img/image2.png"
            ,"/opt/spark/work-dir/notebooks/img/ITESOLogo.png"]

rdd = sc.parallelize(filenames)

```python
text_rdd = rdd.map(extract_text)
```

```python
text_lst = text_rdd.collect()
text_lst
```

```python
spark.stop()
```
