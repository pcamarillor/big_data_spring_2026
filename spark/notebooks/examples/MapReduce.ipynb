```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder \
.appName("1st Spark Session") \
.master("local[*]") \
.config("spark.ui.port","4040") \
.getOrCreate()

```

```python
sc=spark.sparkContext
```

```python
numbers = [1, 43, 5, 24]

rdd= sc.parallelize(numbers)
```

```python
def squares(n):
    return n**2

squares_rdd = rdd.map(squares) # Map phase
r= squares_rdd.collect() # Reduce phase

r
```

```python
data=[
    ('apple',1),
    ('banana',2),
    ('apple',3),
    ('orange',4),
    ('banana',1),
    ('orange',2)
]

rdd_data=sc.parallelize(data)
result=rdd_data.reduceByKey(lambda a,b: a+b)
r=result.collect()      
r
```

```python
grpupped_rdd=rdd_data.groupByKey()
group=grpupped_rdd.collect()

for k,v in group:
    print(f"k: {k}: v{list(v)}")
```

```python
words = ['apple','banana','apple','orange','banana','apple','orange','orange']
words_rdd=sc.parallelize(words)
counter= words_rdd.countByValue()
counter
```
