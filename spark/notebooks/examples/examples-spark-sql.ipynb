## Spark SQL

```python
import pcamarillor.spark_utils as su
spark = su.create_spark_session("spark://spark-master:7077", "spark-sql")
spark.sparkContext
```

```python
from pyspark.sql.types import StringType, StructField, StringType, IntegerType, StructType

data = [('bob', 25), ('alice', 23)]
schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

data_df = spark.createDataFrame(data, schema)
data_df
```
