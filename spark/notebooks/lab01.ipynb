```python
def getUniqueStudents(collabs):
    result = []
    for collab in collabs:
        result.extend(collab)
    return set(result)

collabs = [[1,2,3],[3,4],[1,3,8]]
print(getUniqueStudents(collabs))
```

```python
import itertools

def getNumberOfCollabs(collabs):
    collab_tracker = {}
    for collab in collabs:
        combinations = itertools.combinations(collab, 2)
        for combination in combinations:
            if tuple(combination) in collab_tracker:
                collab_tracker[tuple(combination)] += 1
            else:
                collab_tracker[tuple(combination)] = 1
    return collab_tracker

collabs = [[1,2,3],[3,4],[1,3,8]]
print(getNumberOfCollabs(collabs))
```
