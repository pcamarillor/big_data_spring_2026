```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

unique_students = set()

for group in collaborations:
    for student in group:
        unique_students.add(student)

print(unique_students)
```

```python
pair_count = {}

for group in collaborations:
    n = len(group)
    for i in range(n):
        for j in range(i + 1, n):
            a = group[i]
            b = group[j]

            # orden independiente
            pair = tuple(sorted([a, b]))

            if pair in pair_count:
                pair_count[pair] += 1
            else:
                pair_count[pair] = 1

print(pair_count)
```
