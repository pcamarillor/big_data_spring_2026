Student Collaboration Analyzer

Task 2: Unique Students (Sets)

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

unique_students = set()

for group in collaborations:
    for student in group:
        unique_students.add(student)

print(unique_students)
```

Task 3: Collaboration Frequency (Dictionaries)

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

pair_counts = {}

for group in collaborations:
    unique_group = []
    for student in group:
        if student not in unique_group:
            unique_group.append(student)

    for i in range(len(unique_group)):
        for j in range(i + 1, len(unique_group)):
            s1 = unique_group[i]
            s2 = unique_group[j]

            if s1 < s2:
                pair = (s1, s2)
            else:
                pair = (s2, s1)

            if pair in pair_counts:
                pair_counts[pair] += 1
            else:
                pair_counts[pair] = 1

print(pair_counts)
```
