# Activity: Text Extractor with Map-Reduce

```python
import findspark
findspark.init()


from pyspark.sql import SparkSession
from image_utils import extract_text


spark = SparkSession\
    .builder\
    .appName("Map-Reduce")\
    .master("local[*]")\
    .config("spark.ui.port", "4040")\
    .getOrCreate()

sc = spark.sparkContext
```

```python
filenames = [
    "/opt/spark/work-dir/notebooks/img/ITESOLogo.png"
]

rdd = sc.parallelize(filenames)
```

```python
text_rdd = rdd.map(extract_text)

text_list = text_rdd.collect()

text_list
```

```python
sc.stop()
```
