```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"]
]

# Task 2: Unique Students (Sets)
unique_students = set()
for group in collaborations:
    for student in group:
        unique_students.add(student)

print(f"Estudiantes únicos: {unique_students}")

# Task 3: Collaboration Frequency (Dictionaries)
from itertools import combinations

freq_dict = {}
for group in collaborations:
    # Generamos pares únicos dentro de cada grupo (ordenados alfabéticamente)
    pairs = list(combinations(sorted(group), 2))
    
    for pair in pairs:
        if pair in freq_dict:
            freq_dict[pair] += 1
        else:
            freq_dict[pair] = 1

print(f"Frecuencia de colaboración: {freq_dict}")
```
