```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

```python
def get_unique_students(collaborations):
    students = set()
    for group in collaborations:
        for student in group:
            students.add(student)
    return students
```

```python
print(f'Students: {get_unique_students(collaborations)}')
```

```python
def get_collaboration_frequency(collaborations):
    collaboration_frequencies = dict()

    for group in collaborations:
        group.sort()
        for i in range(0, len(group)):
            for j in range(i + 1, len(group)):
                pair = (group[i], group[j])

                if collaboration_frequencies.get(pair, False):
                    collaboration_frequencies[pair] += 1
                else:
                    collaboration_frequencies[pair] = 1

    return collaboration_frequencies
                
```

```python
print(get_collaboration_frequency(collaborations))
```
