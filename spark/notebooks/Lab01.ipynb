You are given a list of collaboration records. Each record represents a group of students who worked together on an assignment.

Each student is identified by a unique ID.

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

# Tasks

## Task 2: Unique Students (Sets)

Return the set of all unique students involved in all collaboration records.

Guidelines:

        Use a set to avoid duplicate student IDs.

        Iterate over all collaboration records.


```python
def unique_students(collaborations):
    uStudents = set()

    for groups in collaborations:
        uStudents.update(groups)
    return uStudents
    
```

```python
print(unique_students(collaborations))
```

## Task 3: Collaboration Frequency (Dictionaries)

Compute how many times each pair of students collaborated together.

Guidelines:

        Each collaboration pair must be order-independent

        (for example, (s0001, s0002) is the same as (s0002, s0001)).

        Use a dictionary to count the number of collaborations per pair.


```python
def collaboration_frequency(collaborations):
    pairs = {}

    for groups in collaborations:
        length_groups = len(groups)
        if  length_groups < 2:
            continue
        for i in range(length_groups):
            for j in range(i + 1, length_groups):
                first =  groups[i]
                second = groups[j]
                if first > second:
                    pairs_key = (second, first)
                else:
                    pairs_key = (first, second)
                if pairs_key in pairs:
                    pairs[pairs_key] += 1
                else:
                    pairs[pairs_key] = 1
    return pairs
        
    
    
```

```python
collaboration_frequency(collaborations)
```
