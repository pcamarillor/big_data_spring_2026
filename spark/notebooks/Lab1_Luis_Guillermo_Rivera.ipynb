```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

print(collaborations)
```

We created the set and iterate over all the students to add them in the set

```python

def students_as_a_set(collaborations):
    students = set()

    for collaboration in collaborations:
        for student in collaboration:
            students.add(student)

    return students

print(students_as_a_set(collaborations))
```

First sort the collaborations in order to simplify the tuples, then iterate over the combinations and check if it exists in the dict

```python
def collaboration_count(collaborations):
    dict_collaboration = {}
    for collaboration in collaborations:
        sorted_collaboration = sorted(collaboration)
        for i in range(len(sorted_collaboration)):
            for j in range(i+1, len(sorted_collaboration)):
                if ((sorted_collaboration[i], sorted_collaboration[j]) in dict_collaboration):
                    dict_collaboration[(sorted_collaboration[i], sorted_collaboration[j])] += 1
                else:
                    dict_collaboration[(sorted_collaboration[i], sorted_collaboration[j])] = 1
    return dict_collaboration

print(collaboration_count(collaborations))
```
