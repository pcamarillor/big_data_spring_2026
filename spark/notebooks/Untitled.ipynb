```python
from PIL import Image
import pytesseract

img = Image.open("img/ITESOLogo.png")
text = pytesseract.image_to_string(img)

print(text)
```

```python
def extract_text(filename):
    img = Image.open("img/ITESOLogo.png")
    return pytesseract.image_to_string(img)
```

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
            .appName("Map-Reduce-IMG") \
            .master("local[*]") \
            .config("spark.ui.port", "4040") \
            .getOrCreate()

sc = spark.sparkContext

filenames = ["img/ITESOLogo.png", "img/text1.png", "img/text2.png"]

rdd = sc.parallelize(filenames)
```

```python
text_rdd = rdd.map(extract_text)
```

```python
text_lst = text_rdd.collect()
text_lst
```
