```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"]
]
```

```python
def unique_students(collaborations):
    students = set()

    for group in collaborations:
        for student in group:
            students.add(student)

    return students
```

```python
def collaboration_frequency(collaborations):
    freq = {}

    for group in collaborations:
        n = len(group)

        for i in range(n):
            for j in range(i + 1, n):
                a = group[i]
                b = group[j]

                if a < b:
                    pair = (a, b)
                else:
                    pair = (b, a)

                if pair in freq:
                    freq[pair] += 1
                else:
                    freq[pair] = 1

    return freq
```

```python
print("Unique students:")
print(unique_students(collaborations))
print("\nCollaboration frequency:")
print(collaboration_frequency(collaborations))
```
