```python
collaborations = [
["s0001", "s0002", "s0003"],
["s0002", "s0003"],
["s0001", "s0004"],
["s0003", "s0004", "s0005"],
["s0001", "s0002"],
["s0005"],
]
"Task 2: Unique Students (Sets) "
"Return the set of all unique students involved in all collaboration records. "
"Guidelines: Use a set to avoid duplicate student IDs. "
"Iterate over all collaboration records."
def unique_students():
    unique_ids = set()
    for record in collaborations:
        for student_id in record:
            unique_ids.add(student_id)
    print("Unique Students:", unique_ids)



unique_students()
```

```python
"Task 3: Collaboration Frequency (Dictionaries)"
"Compute how many times each pair of students collaborated together."
"Guidelines:"
"Each collaboration pair must be order-independent"
"(for example, (s0001, s0002) is the same as (s0002, s0001))."
"Use a dictionary to count the number of collaborations per pair."

collaborations = [
["s0001", "s0002", "s0003"],
["s0002", "s0003"],
["s0001", "s0004"],
["s0003", "s0004", "s0005"],
["s0001", "s0002"],
["s0005"],
]

def collaboration_frequency():
    dict_freq = {}
    for record in collaborations:
        for i in range(len(record)):
            for j in range(i + 1, len(record)):
                pair = tuple(sorted((record[i], record[j])))
                if pair in dict_freq:
                    dict_freq[pair] += 1
                else:
                    dict_freq[pair] = 1
    print("Collaboration Frequency:", dict_freq)

collaboration_frequency()
```
