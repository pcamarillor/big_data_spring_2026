You are given a list of collaboration records. Each record represents a group of students who worked together on an assignment.

Each student is identified by a unique ID.

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

Unique Students (Sets)

```python
unique_students = set()

for group in collaborations:
    for student in group:
        unique_students.add(student)

print("Unique students:", unique_students)      
    
```

Collaboration Frequency (Dictionaries)

```python
frequency = {}

for group in collaborations:
    # Si es 0 o 1 stud no hay pares
    if len(group) < 2:
        continue

    # Ordenamos para que la pareja sea "order-independent"
    stud_in_order = sorted(group)

    # Generamos todas las parejas posibles dentro del grupo
    for i in range(len(stud_in_order)):
        for j in range(i + 1, len(stud_in_order)):
            pair = (stud_in_order[i], stud_in_order[j])
            frequency[pair] = frequency.get(pair, 0) + 1

print("Pair collaboration counts:")
for pair, count in frequency.items():
    print(pair, "=>", count)
```
