You are given a list of collaboration records. Each record represents a group of students who worked together on an assignment.

**Task 2: Unique Students (Sets)**

Return the set of all unique students involved in all collaboration records.

Guidelines:
Use a set to avoid duplicate student IDs.
Iterate over all collaboration records.

```python
for x in collaborations:
    print(x)
```

```python
uniqueStudents_set = set() 

def getUniqueStudents(list):
    for x in list:
        for y in x:
            uniqueStudents_set.add(y)

getUniqueStudents(collaborations)
        
```

```python
print(uniqueStudents_set)
```

**Task 3: Collaboration Frequency (Dictionaries)**

Compute how many times each pair of students collaborated together.

Guidelines:
Each collaboration pair must be order-independent
(for example, (s0001, s0002) is the same as (s0002, s0001)).
Use a dictionary to count the number of collaborations per pair.

```python

collaborationFrequency_dic = {}

def getCollaborationFrequency(set):
    
    for group in collaborations:

        #Obtain all posible pairs
        for i in range(len(group)):
            for j in range(i+1, len(group)):
                student0 = group[i]
                student1 = group[j]

                #Organize the pait
                pair = tuple(sorted([student0, student1]))

                #+1 to the corresponding pair
                if pair in collaborationFrequency_dic:
                    collaborationFrequency_dic[pair] += 1
                else:
                    collaborationFrequency_dic[pair] = 1

getCollaborationFrequency(uniqueStudents_set)

print(collaborationFrequency_dic)
    
```
