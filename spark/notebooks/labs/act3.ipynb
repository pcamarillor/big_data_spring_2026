```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Map-Reduce-IMG") \
    .master("local[*]") \
    .config("spark.ui.port","4040") \
    .getOrCreate()
```

```python
from PIL import Image
import pytesseract
```

```python
img = Image.open("../img/ITESOLogo.png")
text = pytesseract.image_to_string(img)
text
```

```python
def extractText(filename):
    img = Image.open(filename)
    return pytesseract.image_to_string(img)
```

```python
sc = spark.sparkContext
files = ["/opt/spark/work-dir/notebooks/img/ITESOLogo.png",
         "/opt/spark/work-dir/notebooks/img/img1.png",
         "/opt/spark/work-dir/notebooks/img/img2.png"]

rdd = sc.parallelize(files)
```

```python
rdd_map = rdd.map(extractText)
```

```python
result = rdd_map.collect()
result
```

```python
sc.stop()
```
