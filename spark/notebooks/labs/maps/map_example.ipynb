```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
            .appName("Map-Reduce-IMG") \
            .master("local[*]") \
            .config("spark.ui.port", "4040") \
            .getOrCreate()

spark = spark.sparkContext
```

```python
import os

os.listdir("notebooks/img")
```

```python
images = os.listdir("notebooks/img")

for i in range(len(images)):
    images[i] = "notebooks/img/"+images[i]
    print(images[i])
```

```python
rdd = spark.parallelize(images)
rdd 
```

```python
from maps_utils import extract_text

for i in rdd.collect():
    print(extract_text(i))
```
