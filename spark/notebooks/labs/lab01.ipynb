```python
#You are given a list of collaboration records. 
#Each record represents a group of students who worked together on an assignment.
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

```python
collaborators = set()
collaborationtsFrequency = {}

for collaboration in collaborations:
    for i in range(len(collaboration)):

        #Task 2: Unique Students (Sets)
        #Return the set of all unique students involved in all collaboration records.
        collaborators.add(collaboration[i])

        #Task 3: Collaboration Frequency (Dictionaries)
        #Compute how many times each pair of students collaborated together.
        for j in range(i + 1, len(collaboration)):
            # Create a tuple for each posible pair in every collaboration list
            pair = tuple(sorted([collaboration[i], collaboration[j]]))
            # Increment collaboration value +1, if it was not in the dict, it will get 0 from .get()
            # and Add 1 to that value.
            collaborationtsFrequency[pair] = collaborationtsFrequency.get(pair, 0) + 1


print("Task 2: Unique Students (Sets):", sorted(collaborators))
print("\nTask 3: Collaboration Frequency (Dictionaries):")
for pair, times in collaborationtsFrequency.items():
    print(f"{pair}: {times} Collaborations")




```
