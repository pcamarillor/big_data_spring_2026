```python
from spark_utils import SparkUtils
spark_url = "spark://spark-master:7077"
app_name = "Spark lab 4"
su = SparkUtils(spark_url, app_name)
```

```python
!pip install pyspark
from pyspark.sql.functions import get_json_object
json_schema = SparkUtils.generate_schema([("agency_id", "int"), ("agency_info", "string")])


agencies_df = su._spark \
                .read \
                .option("header", "true") \
                .schema(json_schema) \
                .csv("/opt/spark/work-dir/data/agencies/")

#agencies_df.printSchema()
#agencies_df.show(5,truncate = False)

# At this part, I already have an initial dataframe, but the content of this one has only the id of the object and its information in JSON format;
# Therefore, I need to use get_json_object to extract each column from the JSON. This will result in the correct content of the dataframe. I'll do it for each dataframe. 

agencies_df = agencies_df.withColumn("agency_name", get_json_object(agencies_df.agency_info, "$.agency_name"))

agencies_df = agencies_df.withColumn("city", get_json_object(agencies_df.agency_info, "$.city"))


#agencies_df.show()


```

```python
json_schema = SparkUtils.generate_schema([("city", "int"), ("customer_info", "string")])


customers_df = su._spark \
                .read \
                .option("header", "true") \
                .schema(json_schema) \
                .csv("/opt/spark/work-dir/data/brands/")

"""customers_df.printSchema()
customers_df.show(5,truncate = False)"""


customers_df = customers_df.withColumn("customer_name", get_json_object(customers_df.customer_info, "$.customer_name"))

customers_df = customers_df.withColumn("country", get_json_object(customers_df.customer_info, "$.country"))

#customers_df.show()
```

```python
json_schema = SparkUtils.generate_schema([("car_id", "int"), ("customer_info", "string")])


customers_df = su._spark \
                .read \
                .option("header", "true") \
                .schema(json_schema) \
                .csv("/opt/spark/work-dir/data/cars/")

"""customers_df.printSchema()
customers_df.show(5,truncate = False)"""

customers_df = customers_df.withColumn("customer_name", get_json_object(customers_df.customer_info, "$.customer_name"))

customers_df = customers_df.withColumn("city", get_json_object(customers_df.customer_info, "$.city"))

customers_df = customers_df.withColumn("age", get_json_object(customers_df.customer_info, "$.age"))

customers_df.show()
```

```python
json_schema = SparkUtils.generate_schema([("customer_id", "int"), ("customer_info", "string")])


customers_df = su._spark \
                .read \
                .option("header", "true") \
                .schema(json_schema) \
                .csv("/opt/spark/work-dir/data/customers/")

"""customers_df.printSchema()
customers_df.show(5,truncate = False)"""


customers_df = customers_df.withColumn("customer_name", get_json_object(customers_df.customer_info, "$.customer_name"))

customers_df = customers_df.withColumn("city", get_json_object(customers_df.customer_info, "$.city"))

customers_df = customers_df.withColumn("age", get_json_object(customers_df.customer_info, "$.age"))

#customers_df.show()
```

```python
json_schema = SparkUtils.generate_schema([("rental_id", "int"), ("rental_info", "string")])


rentals_df = su._spark \
                .read \
                .option("header", "true") \
                .schema(json_schema) \
                .csv("/opt/spark/work-dir/data/rentals/")

"""rentals_df.printSchema()
rentals_df.show(5,truncate = False)"""


rentals_df = rentals_df.withColumn("car_id", get_json_object(rentals_df.rental_info, "$.car_id"))

rentals_df = rentals_df.withColumn("customer_id", get_json_object(rentals_df.rental_info, "$.customer_id"))

rentals_df = rentals_df.withColumn("agency_id", get_json_object(rentals_df.rental_info, "$.agency_id"))

#rentals_df.show()
```

```python
#well, in this part I´m gonna generate one single dataframe using join operations, and only one action; because at this part of the code
# I already have each dataframe with its corresponding columns  


df_final = rentals_df.join(cars_df, on="car_id") \
                     .join(agencies_df, on="agency_id") \
                     .join(customers_df, on="customer_id") \
                     .select(
                         "rental_id",
                         "car_name",
                         "agency_name",
                         "customer_name"
                     )

df_final.show()



```
