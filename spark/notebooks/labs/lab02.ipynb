# Lab 02: Bank Account Manager

Problem Statement
Create a BankAccount class and add it in your own module (spark/src/<your module>/).
The class should use a dictionary to store operations (functions) as values and their names (e.g., “deposit”, “withdraw”, “get_balance”) as keys.
Dynamically select and execute operations based on user input.


## Example Usage

```python
from SantiagoEli_utils import BankAccount


account = BankAccount(1000)

account.process["deposit"](13)
account.process["withdraw"](717)

print(f"Current balance: ${account.process['get_balance']()}")
account.display()  
```
