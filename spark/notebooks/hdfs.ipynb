```python
from sebastianM.spark_utils import SparkUtils
from pyspark.sql.functions import get_json_object, col

su = SparkUtils("Lab 04", "spark://spark-master:7077")

df_agencies = su._spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("../data/car_service/agencies/agencies.csv")

df_brands = su._spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("../data/car_service/brands/brands.csv")

df_cars = su._spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("../data/car_service/cars/cars.csv")

df_customers = su._spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("../data/car_service/customers/customers.csv")

df_rentals = su._spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("../data/car_service/rentals/rentals.csv")

df_agencies.show()
```

```python
df_agencies.write.mode("overwrite") \
    .partitionBy("city") \
    .parquet("../data/car_service_parquet/agencies")

df_brands.write.mode("overwrite") \
    .partitionBy("country") \
    .parquet("../data/car_service_parquet/brands")

df_cars.write.mode("overwrite") \
    .partitionBy("brand_id") \
    .parquet("../data/car_service_parquet/cars")

df_customers.write.mode("overwrite") \
    .partitionBy("city") \
    .parquet("../data/car_service_parquet/customers")

df_rentals.write.mode("overwrite") \
    .partitionBy("agency_id") \
    .parquet("../data/car_service_parquet/rentals")
```

```python
df_agencies.printSchema()
```
