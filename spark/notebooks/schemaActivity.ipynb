```python
from SparkInit import init_spark

spark = init_spark("Schema Activity")
```

```python
from SchemaUtils import SchemaUtils 

schema_definition = [
    ("machine", "string"),
    ("date", "date"),
    ("temp", "float"),
]
schema = SchemaUtils().create_schema(schema_definition)
if schema is None:
    print("Error creating schema")
else:
    print(schema)
```

```python
from datetime import date, datetime

factory_data = [
    ("M001", datetime(2026, 1, 26, 8, 10, 0), 75.3),
    ("M002", datetime(2026, 1, 26, 8, 10, 0), 68.7),
    ("M001", datetime(2026, 1, 26, 8, 10, 0), 76.1),
    ("M003", datetime(2026, 1, 26, 8, 15, 0), 72.4),
    ("M002", datetime(2026, 1, 26, 8, 20, 0), 69.8),
    ("M001", datetime(2026, 1, 26, 8, 25, 0), 77.5),
    ("M003", datetime(2026, 1, 26, 8, 30, 0), 73.2),
    ("M002", datetime(2026, 1, 26, 8, 35, 0), 70.1),
    ("M001", datetime(2026, 1, 26, 8, 40, 0), 78.0),
    ("M003", datetime(2026, 1, 26, 8, 45, 0), 74.6),
]
```

```python
dataframe = spark.createDataFrame(factory_data, schema)
dataframe.show()
```

```python
from pyspark.sql.functions import avg, max, min, count, col
```

```python
avg_temp = dataframe.groupBy("machine").agg(
        avg("temp").alias("avg_temp")
    )
avg_temp.show()
```

```python
min_max_temp = dataframe.groupBy("machine").agg(
    max("temp").alias("max_temp"),
    min("temp").alias("min_temp")
)
min_max_temp.show()
```

```python
high_temp = dataframe.filter(col("temp") > 75)
high_temp.show()
```

```python
readings_count = dataframe.groupBy("machine").agg(
    count("temp").alias("count")
)
readings_count.show()
```

```python
hottest = dataframe.orderBy(col("temp"), ascending = False).limit(1)
hottest.show()
```
