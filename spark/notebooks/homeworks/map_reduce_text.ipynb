```python
from mapReduce import extract_text

import findspark
findspark.init()

from pyspark.sql import SparkSession

# we created in local because we just installed tesseract in the notebook container, if we want to use it in a cluster we need to install tesseract in all the nodes of the cluster
spark = SparkSession.builder \
        .appName("Map Reduce IMG Text Extractor") \
        .master("local[*]") \
        .config("spark.ui.port", "4040") \
        .getOrCreate()

sc = spark.sparkContext
```

```python
filenames = [
    '/opt/spark/work-dir/notebooks/img/text1.png',
    '/opt/spark/work-dir/notebooks/img/text2.png',
    '/opt/spark/work-dir/notebooks/img/text3.png',
    '/opt/spark/work-dir/notebooks/img/text4.png',
    '/opt/spark/work-dir/notebooks/img/text5.png'
]

rdd_filenames = sc.parallelize(filenames)
text_rdd = rdd_filenames.map(extract_text)

text_list = text_rdd.collect()
text_list
```

```python
spark.sparkContext.stop()
```
