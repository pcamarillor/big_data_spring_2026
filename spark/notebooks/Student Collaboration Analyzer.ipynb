Unique Students (Sets)

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
unique_students = set()
for group in collaborations:
    unique_students.update(group)
print("Unique students:", unique_students)
```

Collaboration Frequency (Dictionaries)

```python
from itertools import combinations
pairs = {}
for group in collaborations:
    for pair in combinations(group, 2):
        pair = tuple(sorted(pair))
        pairs[pair] = pairs.get(pair, 0) + 1
print("Pair of students:",pairs)
```
