```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]
```

```python
#Task 1
UniqueStudents = set()

for x in collaborations:
    for student in x:
        UniqueStudents.add(student)

UniqueStudents
```

```python
#Task 2
from itertools import combinations

StudentCollabCount = {}

for x in UniqueStudents:
    StudentCollabCount[x] = 0
    
for y in combinations(UniqueStudents, 2):
    StudentCollabCount[y[0]] += 1
        
StudentCollabCount
```
