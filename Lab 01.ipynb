**Task 2: Unique Students (Sets)**

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

mi_set = set()

for i in range(len(collaborations)):
    for j in range(len(collaborations[i])):
        mi_set.add(collaborations[i][j])

print('Students: ', mi_set)
```

**Task 3: Collaboration Frequency (Dictionaries)**

```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

mi_dic = {}

for i in range(len(collaborations)):
    collaborations[i].sort()  
    
    for j in range(len(collaborations[i])):
        for k in range(j + 1, len(collaborations[i])):
            pair = (collaborations[i][j], collaborations[i][k])
            
            if pair in mi_dic:
                mi_dic[pair] += 1
            else:
                mi_dic[pair] = 1

print("Students collabs: ", mi_dic)
```
