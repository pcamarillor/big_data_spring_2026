```python
collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

#Task two 
# In this part, I created a second list to store the elements of the list of list fromn collaborations
c = []
#print(collaborations)

for element in collaborations :
    for elements in element:
        c.append(elements)

#print(c)

#here I created the set

unique_students = set()

# In this part, I added each element of the second list to the set 

for element in c:
    unique_students.add(element)

# This print is only to check the content of the set
print("Unique students ID:\n",unique_students)
```

```python
#Task 3

collaborations = [
    ["s0001", "s0002", "s0003"],
    ["s0002", "s0003"],
    ["s0001", "s0004"],
    ["s0003", "s0004", "s0005"],
    ["s0001", "s0002"],
    ["s0005"],
]

# First, I declared the dictionary for the frequency table use 

dictionary = {}


#Now in this part I sorted the elements of each list, 

for lists in collaborations:
     for i in range(0,len(lists),1):
         for j in range(i+1,len(lists),1):
             first_student = lists[i]
             second_student = lists[j]
             tup = tuple(sorted([first_student,second_student]))

             if tup in dictionary:
                # I write this if to control if tup is already in the dictionary
                # This is because in that case it's necessary to add 1 to the current count
                dictionary[tup] += 1
             else:
                # On the other hand, I only set it to 1, because it's the first time that the current pair appears
                # If it's the first time seeing this pair, set it to 1
                dictionary[tup] = 1
         
#print(dictionary)

#This for is for printing the pairs registered in the table frequency (dictionary)
#Also, I needed to include .item() (an exclusive method of dictionary collection to access each key, value)
for pair, count in dictionary.items():
    print(f"{pair}: {count}")








```
